## Randomness of pi digits distribution

---

What if there is some kind of sacred pattern in pi numbers? We can find out it by just analsying the frequencies of numbers. Also, to fully prove that, we need also to find out whether there is any pattern in sequences' distribution, which, obviously, is more interesting to investigate.

#### Stuff for plots and not only

```python
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
%matplotlib inline
import numpy as np
```

```python
# Set fixed size of graphs
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

#### Load the dataset

```python
with open ( "pi_digits.txt", "r") as file:
    pi = file.read().replace('\n', '')

    
print(len(pi))
print(type(pi))
print(pi[15::-1][::-1]+"...")
```

#### Count every digit frequency

```python
dig_count = [0]*11

for digit in pi[2::]:
    dig_count[int(digit)]+=1
    dig_count[10]+=1
```

```python
'\n' in pi
```

```python
print("Digit count in pi:")
print("Number\tCount")
for i in range(0,10):
    print(str(i) + "\t" + str(dig_count[i]))

print("Total\t" + str(dig_count[10]))
```

#### Pi digits distribution

```python
print("Digit distribution in pi:")
print("Number\tPart(%)")
for i in range(0,10):
    print(str(i) + "\t" + str(dig_count[i]/dig_count[10]*100)[:6]+"%")
```

#### Pi-chart

```python
plt.pie(dig_count[:-1], 
        labels=list(range(0,10)), 
        autopct='%1.1f%%',
        shadow=False,
        labeldistance=1.05,
        startangle=20,
        colors=mcolors.TABLEAU_COLORS)
plt.axis('equal')

font = {'weight' : 'bold',
        'size'   : 18}
plt.rc('font', **font)

sizes= [1500]*10

plt.title("Pi digits distribution")
plt.legend()
plt.show()
```

### Small conclusion
Though digits of pi aren't distributed absolutely equally, their distribution is fair enough to be considered equal.

## Pairs' frequency analysis
---

#### Explanation
My theory is that pi's sequence of digits may have pattern in their changes, so what I call pairs' analysis is analysis of n'th and n+1'th number as a complete one like (pi(n),pi(n-1)). Maybe pattern is in here? 

#### Pairs count

```python
pi_list = list(pi[2::])
for i in range(0, len(pi_list)):
    pi_list[i]=int(pi_list[i])

pair_count = [0]*101
    
for i in range(0, len(pi_list)-1):
    pair_count[pi_list[i]*10+pi_list[i+1]]+=1
    pair_count[100]+=1
```

```python
print("Pairs' count in pi:")
print("Pair\tCount")

delimiter = " - "
for i in range(0,len(pair_count)-1):
    if (len(str(i))>1):
        pair_str = str(i)[0] + delimiter + str(i)[1]
    else:
        pair_str = "0" + delimiter + str(i)[0]
    print(pair_str + "\t" + str(pair_count[i]))

print("Total\t" + str(pair_count[100]))
```

#### Pairs distribution

```python
print("Pairs distribution in pi:")
print("Pair\tPart(%)")
delimiter = " - "
for i in range(0,len(pair_count)-1):
    if (len(str(i))>1):
        pair_str = str(i)[0] + delimiter + str(i)[1]
    else:
        pair_str = "0" + delimiter + str(i)[0]
    print(pair_str + "\t" + str(pair_count[i]/pair_count[100]*100)[:6]+"%")
```

```python
pairs_str_list = []
delimiter = ""
for i in range(0,len(pair_count)-1):
    if (len(str(i))>1):
        pair_str = str(i)[0] + delimiter + str(i)[1]
    else:
        pair_str = "0" + delimiter + str(i)[0]
    pairs_str_list.append(pair_str)
```

```python
plt.pie(pair_count[:-1], labels=pairs_str_list, 
        autopct='%1.f%%', shadow=False,
        labeldistance=1.07, startangle=90,
        colors=mcolors.TABLEAU_COLORS, pctdistance=0.9,
        counterclock=False)


font = {'weight' : 'bold',
        'size'   : 10}
plt.rc('font', **font)

plt.title("Pi pairs distribution")
plt.show()
```

#### Small conclusion

Even the order of numbers is distributed relatively equal in pi digits. Random enough. What about triplets?

### Triplets' analysis
---

```python
pi_list = list(pi[2::])
for i in range(0, len(pi_list)):
    pi_list[i]=int(pi_list[i])
    
pi_triplets = []

for i in range(0, len(pi_list)-2):
    pi_triplets.append([pi_list[i], pi_list[i+1], pi_list[i+2]])
```

#### Triplets count

```python
pi_triplets_count = [0]*1001

for i in range(0, len(pi_triplets)):
    pi_triplets_count[pi_triplets[i][0]*100+pi_triplets[i][1]*10+pi_triplets[i][2]*1]+=1
    pi_triplets_count[1000]+=1
    
print(pi_triplets_count[0:10])
```

```python
print("Triplets' count in pi:")
print("Triplet\tCount")

triplet_str_list = []

for i in range(0,1000):
    triplet_str = str(i)
    
    if (len(triplet_str)==1):
        triplet_str = "0" + triplet_str
    
    if (len(triplet_str)==2):
        triplet_str = "0" + triplet_str
    
    triplet_str_list.append(str(triplet_str))
    
    
for i in range(0,1000):
    print(triplet_str_list[i] + "\t" + str(pi_triplets_count[i]))

print("Total\t" + str(pi_triplets_count[1000]))
```

#### Triplets distribution

```python
print("Triplets' distribution in pi:")
print("Triplet\tPart(%)")

for i in range(0,1000):
    print(triplet_str_list[i] + "\t" + str(pi_triplets_count[i]*100/pi_triplets_count[1000])[:6])
```

#### Frequency difference analysis
The fact is that perfect distribution dispersion equals zero, which in our case is obviously not. *Especially* in case of triplets. May thee find the pattern there.

```python
# Calc the average
mean1 = np.array(pi_triplets_count[:-1]).mean()

sum = 0
for i in range(0, len(pi_triplets_count)-1):
    sum += pi_triplets_count[i]
    
mean2 = sum/(len(pi_triplets_count)-1)

print(mean1)
print(mean2)

mean = mean1
```

```python
# Calc the dispersion for each of digits
disp_trip_list = []

for i in range(0, len(pi_triplets_count[:-1])):
    disp_trip_list.append(pi_triplets_count[i]-mean)     # abs is possible
```

```python
print("Dispersion for each triplet")
print("Triplet\tDispersion")

disp_zip = zip(triplet_str_list, disp_trip_list[:])
disp_zip_sorted = sorted(disp_zip, key=lambda tup: tup[1])

print(len(disp_trip_list)==len(disp_zip_sorted))

for i in range(0, len(pi_triplets_count[:-1])):
    print(disp_zip_sorted[i][0] + "\t" + str(disp_zip_sorted[i][1])[:6])
```

```python
print("Dispersion distribuetion for each triplet")
print("shows dispersion to count ratio")
print("how strongly did dispersion help to make this frequentier")
print("Triplet\tDispersion part(%)")

dispart_list = []
for i in range(0, len(disp_trip_list)):
    dispart_list.append(disp_trip_list[i]/mean*100)

disp_zip = zip(triplet_str_list, dispart_list[:])
disp_zip_sorted = sorted(disp_zip, key=lambda tup: tup[1])


for i in range(0, len(pi_triplets_count[:-1])):
    print(disp_zip_sorted[i][0] + "\t" + str(disp_zip_sorted[i][1])[:6])
```

#### Conclusion
Yes, we can definitely say that numbers taken from random positions in pi are random. But that doesn't work well for sequences! *Some sequences are more common than others.* It might be interesting to investigate deeper, using longer sequences, which is an area to research. 

Talking about triplets, the most common one is 654, and the least common is 067. Difference between their counts is **194**, which is insanely huge difference comparing to overall average count (around *1000* of each triplet). Still, I didn't investigate numbers themselves which is also an idea for research.

My hypotesis is that longer sequences of numbers are less random. Still, I should assume the fact that analysis of longer sequences require more data. So I should be using bigger dataset instead for complete analysis.
